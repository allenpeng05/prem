```python
import pandas as pd

#load in data from scrapers
matches = pd.read_csv("matches.csv",index_col=0)
standings = pd.read_csv("standings_2014_15_to_2023_24.csv")


# map missing team names
class MissingDict(dict):
    def __missing__(self, key):
        return key


map_values = {
    "Brighton and Hove Albion": "Brighton",
    "Manchester United": "Manchester Utd",
    "Newcastle United": "Newcastle Utd",
    "Tottenham Hotspur": "Tottenham",
    "West Ham United": "West Ham",
    "Wolverhampton Wanderers": "Wolves",
}
mapping = MissingDict(**map_values)
matches["team"] = matches["team"].map(mapping)

# calculate the previous season
matches["previous_season"] = matches["season"] - 1

# merge the two dfs
merged_df = matches.merge(
    standings.rename(columns={"season": "previous_season", "rank": "previous_rank"}),
    on=["previous_season", "team"],
    how="left",
)
merged_df["promoted_flag"] = merged_df["previous_rank"].isnull().astype(int)
merged_df["previous_rank"] = merged_df["previous_rank"].fillna(-1).astype(int)
```
